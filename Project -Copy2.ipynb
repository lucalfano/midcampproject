```python
def bigfunc(df):

#     df['year'] = pd.DatetimeIndex(df['Date']).year

    # organizing, renaming cols, and attributing points to winner/home
    hometeamgames = df[['Date', 'HomeTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'HST', 'HF', 'HC', 'HY', 'HR']]
    
    hometeamgames['homepoints'] = hometeamgames['FTR'].map({'H': 3, 'D': 1, 'A': 0})
    
#     hometeamgames['year'] = pd.DatetimeIndex(df['Date']).year

    
    homewin = []
    homedraw = []
    homeloss = []
    
    for i in hometeamgames['FTR']:
        if 'A' in i:
            homewin.append(0)
            homedraw.append(0)
            homeloss.append(1)
        elif 'D' in i:
            homewin.append(0)
            homedraw.append(1)
            homeloss.append(0)
        elif 'H' in i:
            homewin.append(1)
            homedraw.append(0)
            homeloss.append(0)
    hometeamgames['homewin'] = homewin
    hometeamgames['homedraw'] = homedraw
    hometeamgames['homeloss'] = homeloss
    
    homegamesresults = hometeamgames.groupby('HomeTeam').sum().reset_index()

    homegamesresults = homegamesresults.rename(columns={'FTHG':'HGS', 'FTAG':'HGC'})
    
    homegamesresults['homeplayed'] = homegamesresults['homewin']+homegamesresults['homedraw']+homegamesresults['homeloss']
    
    
    
    
    
    # organizing, renaming cols, and attributing points to winner/away
    awayteamgames = df[['Date', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'AS', 'AST', 'AF', 'AC', 'AY', 'AR']]
    
    awayteamgames['awaypoints'] = awayteamgames['FTR'].map({'H': 0, 'D': 1, 'A': 3})
    
    awaywin = []
    awaydraw = []
    awayloss = []
    
    for i in awayteamgames['FTR']:
        if 'A' in i:
            awaywin.append(1)
            awaydraw.append(0)
            awayloss.append(0)
        elif 'D' in i:
            awaywin.append(0)
            awaydraw.append(1)
            awayloss.append(0)
        elif 'H' in i:
            awaywin.append(0)
            awaydraw.append(0)
            awayloss.append(1)
    awayteamgames['awaywin'] = awaywin
    awayteamgames['awaydraw'] = awaydraw
    awayteamgames['awayloss'] = awayloss
        
    awaygamesresults = awayteamgames.groupby('AwayTeam').sum().reset_index()
    
    awaygamesresults = awaygamesresults.rename(columns={'FTHG':'AGC', 'FTAG':'AGS'})
    
    awaygamesresults['awayplayed'] = awaygamesresults['awaywin']+awaygamesresults['awaydraw']+awaygamesresults['awayloss']
    
    
    
    
    # back together to find season totals 
    seasonresults = pd.concat([homegamesresults, awaygamesresults], axis=1)
    
#     seasonresults['year'] = pd.DatetimeIndex(hometeamgames['Date']).year
    # agg cols 
    seasonresults['totalpoints'] = seasonresults['homepoints']+seasonresults['awaypoints']
    
    seasonresults['gamesplayed'] = seasonresults['homeplayed'] + seasonresults['awayplayed']
    
    seasonresults['TGS'] = seasonresults['HGS']+seasonresults['AGS']
    
    seasonresults['TGC'] = seasonresults['HGC']+seasonresults['AGC']
    
    seasonresults['GD'] = seasonresults['TGS']-seasonresults['TGC']

    seasonresults = seasonresults.rename(columns={'HomeTeam':'Team'})

    #ratios 
    
    seasonresults['winsagg'] = seasonresults['homewin']+seasonresults['awaywin']
    seasonresults['drawsagg'] = seasonresults['homedraw']+seasonresults['awaydraw']
    seasonresults['lossagg'] = seasonresults['homeloss']+seasonresults['awayloss']
    
#     seasonresults['win%'] = round((seasonresults['winsagg']/seasonresults['gamesplayed']*100),1)
#     seasonresults['draw%'] = round((seasonresults['drawsagg']/seasonresults['gamesplayed']*100),1)
#     seasonresults['loss%'] = round((seasonresults['lossagg']/seasonresults['gamesplayed']*100),1)

    seasonresults = seasonresults[['Team', 'HGS', 'HGC', 'AGS', 'AGC','HS', 'AS', 'HST', 'AST', 'HF','AF','HC','AC', 'HY', 'AY', 'HR', 'AR' , 'totalpoints', 'TGS', 'TGC', 'GD', 'homepoints', 'awaypoints', 'gamesplayed', 'homewin', 'homedraw', 'homeloss', 'awaywin', 'awaydraw', 'awayloss','winsagg','drawsagg','lossagg']]

    return seasonresults
```
